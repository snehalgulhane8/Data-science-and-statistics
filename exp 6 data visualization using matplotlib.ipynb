```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=2*x
```

```python
y
```

line chart

```python
plt.plot(x,y)
plt.title("Line chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

Bar chart

```python
plt.bar(x,y)
plt.title("Bar chart")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.show()
```

```python
plt.bar(x,y,color='red')
```

```python
a=(1,4,5,7,3,6,8,9,4,2)
b=(10,25,55,64,23,56,11,67,45,34)
plt.scatter(a,b)
plt.title("scatter plot")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.show()
```
